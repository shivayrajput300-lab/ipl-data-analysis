```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("IPL.csv", low_memory=False)
print("Data loaded! Shape:", df.shape)
```

```python
top_batsmen = df.groupby('batter')['runs_batter'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(14, 7))
ax = sns.barplot(x=top_batsmen.values, y=top_batsmen.index, palette='flare')

for i, val in enumerate(top_batsmen.values):
    ax.text(val + 50, i, f'{val:,}', va='center', fontsize=11, fontweight='bold')

plt.title('Top 10 Batsmen by Total Runs — IPL 2008 to 2026', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Total Runs', fontsize=13)
plt.ylabel('Batsman', fontsize=13)
plt.tight_layout()
plt.show()
plt.savefig("shivvv.png")
```

```python
team_wins = df[df['match_won_by'].notna()].groupby('batting_team')['match_id'].nunique()

top_teams = team_wins.sort_values(ascending=False).head(8)

plt.figure(figsize=(12, 6))
ax = sns.barplot(x=top_teams.index, y=top_teams.values, palette='crest')

for i, val in enumerate(top_teams.values):
    ax.text(i, val + 1, str(val), ha='center', fontsize=11, fontweight='bold')

plt.title('IPL Team Wins — All Seasons', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Team', fontsize=13)
plt.ylabel('Number of Wins', fontsize=13)
plt.xticks(rotation=20, ha='right')
plt.tight_layout()
plt.show()
```

```python

toss = df.drop_duplicates('match_id')['toss_decision'].value_counts()

plt.figure(figsize=(7, 7))
plt.pie(
    toss.values,
    labels=toss.index,
    autopct='%1.1f%%',
    colors=['c', 'y'],
    startangle=90,
    wedgeprops={'edgecolor': 'white', 'linewidth': 2},
    textprops={'fontsize': 10}
)
plt.title('Toss Decision — Bat vs Field', fontsize=10, fontweight='bold', pad=10)
plt.tight_layout()
plt.show()
```

```python
season_runs = df.groupby(['season', 'match_id'])['runs_total'].sum().reset_index()
season_avg = season_runs.groupby('season')['runs_total'].mean()

plt.figure(figsize=(14, 6))
plt.plot(season_avg.index, season_avg.values, marker='o', color='#E91E63', linewidth=2.5, markersize=7)
plt.fill_between(season_avg.index, season_avg.values, alpha=0.15, color='#E91E63')

for x, y in zip(season_avg.index, season_avg.values):
    plt.text(x, y + 1.5, f'{y:.0f}', ha='center', fontsize=9, fontweight='bold')

plt.title('Average Runs per Match — Season Wise Trend', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Season', fontsize=13)
plt.ylabel('Avg Runs per Match', fontsize=13)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
top_sixers = df[df['runs_batter'] == 6].groupby('batter').size().sort_values(ascending=False).head(10)

plt.figure(figsize=(14, 7))
ax = sns.barplot(x=top_sixers.values, y=top_sixers.index, palette='rocket')

for i, val in enumerate(top_sixers.values):
    ax.text(val + 1, i, str(val), va='center', fontsize=11, fontweight='bold')

plt.title('Top 10 Six Hitters — IPL All Time 💥', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Number of Sixes', fontsize=13)
plt.ylabel('Batsman', fontsize=13)
plt.tight_layout()
plt.show()
```

```python
top_fours = df[df['runs_batter'] == 4].groupby('batter').size().sort_values(ascending=False).head(10)

plt.figure(figsize=(14, 7))
ax = sns.barplot(x=top_fours.values, y=top_fours.index, palette='mako')

for i, val in enumerate(top_fours.values):
    ax.text(val + 1, i, str(val), va='center', fontsize=11, fontweight='bold')

plt.title('Top 10 Four Hitters — IPL All Time 🏏', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Number of Fours', fontsize=13)
plt.ylabel('Batsman', fontsize=13)
plt.tight_layout()
plt.show()
```

```python
top_bowlers = df.groupby('bowler')['bowler_wicket'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(14, 7))
ax = sns.barplot(x=top_bowlers.values, y=top_bowlers.index, palette='magma')

for i, val in enumerate(top_bowlers.values):
    ax.text(val + 0.3, i, str(int(val)), va='center', fontsize=11, fontweight='bold')

plt.title('Top 10 Wicket Takers — IPL All Time 🎯', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Total Wickets', fontsize=13)
plt.ylabel('Bowler', fontsize=13)
plt.tight_layout()
plt.show()
```
